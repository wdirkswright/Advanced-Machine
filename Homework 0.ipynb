```python
import pandas as pd
import matplotlib.pyplot as plt

data = {"City": ["Seattle", "San Francisco", "New York"],
        "Rent": [1800, 2550, 2500],
        "Food": [520, 830, 875],
        "Transportation": [110, 179, 373],
        "Utilities": [250, 250, 220],
        "Misc": [474, 231, 239]
        }

df = pd.DataFrame(data)
```

```python
df["Edu"] = 600
df['Monthly_Cost'] = df['Rent'] + df['Food'] + df['Transportation'] + df['Utilities'] + df['Misc'] +df['Edu']
df["Minimum_Annual_Salary"] = df['Monthly_Cost'] * 12
df['Desired_Salary'] = df['Minimum_Annual_Salary'] * 2

df
```

```python
df.plot(x="City", y=["Minimum_Annual_Salary", 'Desired_Salary'], kind="bar")
plt.ylabel("Annual Salary ($)")
plt.title("Minimum Annual Salary Needed by City")
plt.show()
```
