```python
### import packages
import pandas as pd
import sweetviz as sv
import numpy as np
import matplotlib.pyplot as plt
```

```python
### load dataset
df = pd.read_csv('train.csv')
df.head()
```

```python
### sweetviz report
report = sv.analyze(df)
report.show_html('irrigation_sweetviz_report.html')
```

### Associations

Variables that appear to have strong associations with the response variable irrigation need include soil moisture, temperature, rainfall, wind speed, and crop growth stage. Out of these, soil moisture appeared to have the most significant association. 

Additionally, there doesn't appear to be strong associations between predictors.

```python
### distribution of target variable (irrigation_need)
plt.figure(figsize=(6, 4))
df['Irrigation_Need'].value_counts().plot(kind='bar')
plt.title('Distribution of Irrigation Need')
plt.xlabel('Irrigation Need')
plt.ylabel('Count') 
```

```python
### boxplot of each numeric feature by target variable (irrigation_need)
import seaborn as sns
numeric_features = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_features = [col for col in numeric_features if col != 'id']  # Remove 'id' column

for feature in numeric_features:
    plt.figure(figsize=(9, 6))
    sns.boxplot(data=df, x='Irrigation_Need', y=feature, palette={'Low': '#66c2a5', 'Medium': '#4575b4', 'High': '#d73027'})
    plt.title(f'Boxplot of {feature} by Irrigation Need', fontsize=14, fontweight='bold')
    plt.xlabel('Irrigation Need', fontsize=12)
    plt.ylabel(feature, fontsize=12)
    plt.grid(False)
    plt.tight_layout()
    plt.show()
```

```python
### grouped bar plots of categorical features by target variable
categorical_features = df.select_dtypes(include=['object']).columns.tolist()
categorical_features = [col for col in categorical_features if col != 'Irrigation_Need']

for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    crosstab = pd.crosstab(df[feature], df['Irrigation_Need'])
    crosstab = crosstab[['Low', 'Medium', 'High']]  # Reorder columns
    crosstab.plot(kind='bar', ax=plt.gca(), 
                  color=['#66c2a5', '#8da0cb','#fc8d62'], 
                  edgecolor='black', linewidth=1.2, width=0.8)
    plt.title(f'Count of Irrigation Need by {feature}', fontsize=14, fontweight='bold')
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.legend(title='Irrigation Need', loc='upper right')
    plt.xticks(rotation=45, ha='right')
    plt.grid(False)
    plt.tight_layout()
    plt.show()
```
