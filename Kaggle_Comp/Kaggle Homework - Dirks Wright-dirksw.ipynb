## Bagging & Boosting Notebook

### Modeling Approaches & Discussion

#### Overview of Approaches

This analysis employed two primary ensemble learning techniques to classify irrigation needs:
1. **Bagging (Random Forest)** - An approach that reduces variance by training multiple models on random subsets of data
2. **Boosting (XGBoost)** - An approach that reduces bias by sequentially training models to correct previous errors

#### Models Tested

**Baseline Models (No Hyperparameter Tuning):**
- Random Forest with default parameters (100 estimators)
- XGBoost with default parameters

**Tuned Models (Hyperparameter Optimization):**
- Random Forest with Optuna (10 trials): Optimized n_estimators, max_depth, min_samples_split, min_samples_leaf, max_features
- XGBoost with Optuna (10 trials): Optimized n_estimators, max_depth, learning_rate, subsample, colsample_bytree

#### Performance Metrics Summary

The comprehensive model comparison reveals the following validation set performance metrics:

| Model | CV F1 Score | Val Accuracy | Val F1 (Weighted) | Val F1 (Macro) |
|-------|------------|--------------|-------------------|----------------|
| Random Forest (Baseline) | 0.8562 | 0.9364 | 0.9362 | 0.8912 |
| XGBoost (Baseline) | 0.8605 | 0.9372 | 0.9371 | 0.8983 |
| Random Forest (Tuned) | 0.8603 | 0.9367 | 0.9366 | 0.8924 |
| **XGBoost (Tuned)** | **0.8694** | **0.9391** | **0.9391** | **0.9073** |

#### Leaderboard Performance

| Model | Leaderboard Score | Position | 
|-------|------------------|--------------|
| Random Forest (Tuned) | 0.95489 | 1744 |

The leaderboard scores confirm the validation set results, with XGBoost achieving superior performance on the final test set.

#### Key Findings

**Best Performing Model:** XGBoost (Tuned) achieved the highest performance across all metrics with a validation F1 score of 0.9391 (weighted).

**Improvement from Tuning:**
- Random Forest: +0.0004 F1 points (0.9362 → 0.9366) - **Minimal improvement**
- XGBoost: +0.0020 F1 points (0.9371 → 0.9391) - **Modest improvement**

The improvements from hyperparameter tuning were relatively small, suggesting that the baseline models were already well-configured for this problem. However, XGBoost's tuning led to more meaningful gains than Random Forest's.

#### Bagging vs. Boosting: Comparative Analysis

**Random Forest (Bagging) Results:**
- Baseline Accuracy: 93.64%
- Tuned Accuracy: 93.67%
- CV F1 Mean: 0.8562 → 0.8603 (+0.0041)

**XGBoost (Boosting) Results:**
- Baseline Accuracy: 93.72%
- Tuned Accuracy: 93.91%
- CV F1 Mean: 0.8605 → 0.8694 (+0.0089)

**Bagging vs. Boosting Insights:**

1. **Performance Advantage - Boosting Wins:** XGBoost outperformed Random Forest by ~0.0025 F1 points on validation data (0.9391 vs 0.9366), demonstrating boosting's ability to achieve better predictive accuracy for this multi-class irrigation classification task.

2. **Optimization Responsiveness:** XGBoost showed greater improvement from hyperparameter tuning (+0.0089 CV F1) compared to Random Forest (+0.0041 CV F1), suggesting that boosting models are more sensitive to hyperparameter configurations and can benefit from careful tuning.

3. **Generalization:** Both models achieved similar cross-validation F1 scores in their baseline forms (0.8562 vs 0.8605), but XGBoost's tuned version achieved a notably higher CV F1 score (0.8694), indicating better generalization after optimization.

4. **Class-Specific Performance:** The macro F1 scores (which weigh all classes equally) show XGBoost (Tuned) at 0.9073 vs. Random Forest (Tuned) at 0.8924, indicating XGBoost better handles class imbalance and provides more balanced performance across the three irrigation need categories.

#### Feature Importance Comparison

Both models identified **Crop_Growth_Stage** as the most important feature, followed by **Soil_Moisture**. This alignment suggests these features are genuinely predictive of irrigation needs. However:
- **Random Forest** emphasizes **Wind_Speed_kmh** and **Temperature_C** in its top 5
- **XGBoost** emphasizes **Mulching_Used**, highlighting different aspects of the feature space


#### Conclusion

While both bagging and boosting approaches achieved strong performance (93.6%+ accuracy), **XGBoost demonstrated superior performance** with a validation accuracy of 93.91% and a weighted F1 score of 0.9391. The improvements from hyperparameter tuning were modest (0.2-0.9%), but XGBoost's greater responsiveness to tuning and superior final performance make it the recommended model for this irrigation classification task. The similarity in feature importance between models validates that the key predictive signals have been captured.
