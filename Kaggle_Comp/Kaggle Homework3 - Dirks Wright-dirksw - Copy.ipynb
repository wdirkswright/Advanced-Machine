### Modeling Approaches

This analysis employed two primary boosting learning techniques to classify irrigation needs:
1. Gradient Boosting, through a gradient boosting classifier, which attempted to reduce prediction error by sequentially training multiple models that correct the mistakes of previous ones
2. Categorical Boosting through a cat boost classifier, which attempted to reduce prediction bias by sequentially training models that correct previous errors while effectively handling categorical features

-   Catboost is a specialized form of gradient boosting that natively handles categorical features, Gradient boosting requires manual preprocessing 

#### Models Tested

Baseline Models (No Hyperparameter Tuning):
- GradientBoost with default parameters
- CatBoost with default parameters

Tuned Models: 3 sets of Values for Hyperparameter Optimization:
-   Gradient Boosting (GBM Tuned): Increased n_estimators (250), reduced learning_rate (0.05), moderate depth (3), and subsampling (0.8) to balance bias and variance
-   Gradient Boosting (GBM Conservative): Higher n_estimators (400), very low learning_rate (0.02), shallow trees (depth=2), and stronger regularization to test underfitting.
-   Gradient Boosting (GBM Aggressive): Fewer trees (50), very low learning_rate (0.01), deeper trees (depth=5), and no subsampling to explore overfitting.
-   CatBoost (Baseline): Moderate iterations (150), standard learning_rate (0.10), and depth (6) using multiclass loss as a baseline configuration.
-   CatBoost (Tuned): Increased iterations (300), lower learning_rate (0.05), deeper trees (depth=8), and mild regularization for smoother learning.
-   CatBoost (Conservative): High iterations (500), very low learning_rate (0.02), shallow depth (3), and stronger regularization to test underfitting.
-   CatBoost (Aggressive): Moderate iterations (200), very low learning_rate (0.01), deep trees (depth=10), and reduced regularization to test overfitting.

The improvements from hyperparameter tuning were very small, indicating that both baseline models were already highly optimized for this dataset. While tuning provided slight gains, neither Gradient Boosting nor CatBoost showed substantial performance increases, suggesting limited sensitivity to further hyperparameter adjustments in this case.

#### Hyperparameter inputs

Testing different hyperparameter configurations showed that model performance was relatively stable, with only marginal improvements from tuning. The “tuned” settings for both Gradient Boosting and CatBoost provided slight gains, while the conservative configurations generally reduced performance due to underfitting, and the aggressive configurations also performed worse, likely due to overfitting or unstable learning. Overall, this suggests the baseline models were already close to optimal, and that extreme changes to hyperparameters tended to harm performance rather than improve it.

#### Conclusion

Overall, both Gradient Boosting and CatBoost demonstrated strong and consistent performance across all configurations, with only marginal gains achieved through hyperparameter tuning. The tuned models slightly outperformed their baselines, but the improvements were minimal, indicating that the default settings were already well suited for the problem. More extreme configurations, either overly conservative or aggressive, generally led to decreased performance, highlighting the importance of balanced parameter choices. However, neither of the models performed better on the final test set than XGBoost did last time. In conclusion, while tuning provided minor refinements, the results suggest that model selection and data quality had a greater impact than extensive hyperparameter optimization in this case.
